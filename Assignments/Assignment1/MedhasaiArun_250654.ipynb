## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
marks = []

for i in range(10):
    marks.append(int(input("Enter the marks for 10 students:")))
average= sum(marks)/10
highest= max(marks)
lowest = min(marks)
count=0
print("Highest marks :", highest)
print("Lowest marks :", lowest)
print("Average marks :", average)
for score in marks:
    if score>average: count+=1
print("No. of students considered as above average :", count)
gradedist={"A":0,"B":0,"C":0,"D":0}
for score in marks:
    if score>=90: gradedist["A"]+=1
    elif score>=75 and score<90: gradedist["B"]+=1
    elif score>=60 and score<75: gradedist["C"]+=1
    else : gradedist["D"]+=1
print("GRADE DISTRIBUTION :", gradedist)
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python

 ["R","R","G"],

]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
face = [
    ["R","R","G"],
    ["W","R","G"],
    ["B","B","Y"]
]
 
for row in face:
    for value in row:
        print(value, end=" ")
    print()      
counts={}
for row in face:
    for value in row:
        if value in counts:
            counts[value]+=1
        else: counts[value]=1
print(counts)
for row in face:
    for index in range(len(row)):
        if row[index]=="R": row[index]="X" 
print(face)        

highestcount=0        
mostcomcolor=""
for color in counts:
    if counts[color]>highestcount:
        highestcount= counts[color]
        mostcomcolor=color
print("The most common color is:", mostcomcolor)
nrows=0
for row in face:
    if row[0]==row[1] and row[0]==row[2]:
        nrows+=1
if nrows>0:print("YES")
else: print("NO")



````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
sentence=input("Enter a sentence:")
nofchar= len(sentence)
print("Total characters:", nofchar)
nofwords=0;
for char in sentence:
    if char==" ": nofwords+=1
if nofwords>0: 
    print("No. of words :", nofwords+1)    

vowels="aeiouAEIOU"
nofvowels=0
for char in sentence:
    if char in vowels:
        nofvowels+=1
    
print("No. of vowels :",nofvowels)    
dict={}
for char in sentence:
    if char in dict: dict[char]+=1
    else: dict[char]=1
print(dict)
mostfreqc=""
highestcount=0
for char in dict:
    if dict[char]>highestcount: 
        highestcount=dict[char]
        mostfreqc=char
print("The most frequent character is :",mostfreqc)
modstr=""
for char in sentence:
    if char!=" ":
        modstr=modstr+char.lower()
revstr=""
for char in sentence:
    if char!=" ":
        revstr=char.lower()+ revstr
if revstr==modstr: print("Palindrome : TRUE")
else: print("Palindrome : FALSE")        
        
````
