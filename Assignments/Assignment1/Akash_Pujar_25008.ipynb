## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
# Write your solution here
marks = []
tot, min, max = 0, 101, 0
i = 1
while i<=10:
    mark = int(input("Enter marks of student " + str(i) + ": "))
    marks.append(mark)
    i += 1
    tot+=mark
    if max < mark:
        max = mark
    if min > mark:
        min = mark

avg = tot/(i-1)
abv_avg = 0
for mark in marks:
    if mark > avg:
        abv_avg += 1

grades = {"A": 0, "B": 0, "C": 0, "D": 0}
for mark in marks:
    if mark >= 90:
        grades["A"] += 1
    elif mark >= 75:
        grades["B"] += 1
    elif mark >= 60:
        grades["C"] += 1
    else:
        grades["D"] += 1

print("Total marks =",tot)
print("Highest marks =", max)
print("Lowest marks =", min)
print("Average marks =", avg)
print("Students scoring above average =", abv_avg)
print("Grade distribution =", grades)
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
# Write your solution here
colors = ["W", "Y", "R", "O", "B", "G"]
face = [ ["R", "R", "G"], ["R", "W", "G"], ["B", "B", "Y"] ]

for a in face:
    for b in a:
        if b in colors:
            print(b, end = " ")
        else:
            print("(Invalid)", end = " ")
    print(" ")

freq = {}
for a in face:
    for b in a:
        if b in freq:
            freq[b] += 1
        else:
            freq[b] = 1

print("\nColor Frequencies:")
for color in freq:
    print(color, ":", freq[color])

print(" ")
for a in face:
    for b in a:
        if b in colors:
            if b == "R":
                b = "X"
            print(b, end = " ")
        else:
            print("(Invalid)", end = " ")
    print(" ")

most_common = ""
max_count = 0

for color in freq:
    if freq[color] > max_count:
        max_count = freq[color]
        most_common = color

print("\nMost common color:", most_common)

identical_row = False

for a in face:
    if a[0] == a[1] == a[2]:
        identical_row = True

if identical_row:
    print("\nA row contains all identical colors.")
else:
    print("\nNo row contains all identical colors.")
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
# Write your solution here
sent = input("Enter sentence: ")
char = 0
for ch in sent:
    char+=1

word = 0
prev = ' '
for curr in sent:
    if curr != ' ' and prev == ' ':
        word+=1
    prev = curr

vowels = 0
for ch in sent:
    if ch.lower() in "aeiou":
        vowels+= 1

freq = {}
max_freq = 0
for ch in sent:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

    if freq[ch] > max_freq:
        max_freq = freq[ch]
        max_char = ch


print("Total Characters:", char)
print("Total Words:", word)
print("Total Vowels:", vowels)
print("\nCharacter Frequencies:")
for ch in freq:
    print(ch, ":", freq[ch])
print("\nMost Frequent Character:", max_char)
````
