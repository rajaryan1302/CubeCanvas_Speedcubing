## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
# Write your solution here
marks=[]

for i in range(10)
    mark=float(input(f"Enter marks {i+1}:"))
    marks.append(mark)

lowest = min(marks)
highest = max(marks)
average = sum(marks) / len(marks)
aboveaverage = sum(1 for mark in marks if mark > average)

print("Highest Marks:", highest)
print("Lowest Marks:", lowest)
print("Average Marks:", average)
print("Students Scoring Above Average:", aboveaverage)

gradedistribution = {"A": 0, "B": 0, "C": 0,"D": 0}

for mark in marks:
    if mark >=90:
        grade = "A"
    elif mark >= 75:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "D"

    gradedistribution[grade] +=1

print("\n Grade Distribution")

for grade, count in gradedistribution.items():
    print(f"{grade}: {count}") 
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
# Write your solution here
face = [
    ["R", "R", "G"],
    ["W", "R", "G"],
    ["B", "B", "Y"]
]

print("Original Grid:")
for row in face:
    print(" ".join(row))

frequency = {}

for row in face:
    for colour in row:
        if colour in frequency:
            frequency[colour] += 1
        else:
            frequency[colour] = 1 

print("\nColour Frequencies:")
print(frequency)

for i in range(len(face)):
    for j in range(len(face[i])):
        if face[i][j] == "R":
            face[i][j] = "X"

print("\nAfter replacing R with X:")
for row in face:
    print(" ".join(row))

mostcommon = max(frequency,key=frequency.get)
print("\nMost common colour:",mostcommon)

identicalrow = False

for row in face:
    if row.count(row[0]) == len(row):
        identicalrow = True
        break
    
if identicalrow:
    print("A row contains all identical colours.")
else:
    print("No row contains all the identical colours.")        
````
