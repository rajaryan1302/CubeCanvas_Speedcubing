## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
# Write your solution here
#taking input of 10 students
marks=[]
for i in range(10):
  mark = float(input(f"Enter marks {i + 1}: "))
  marks.append(mark)
#Calulations
highest=max(marks)
lowest=min(marks)
average=sum(marks)/ len(marks)

count=0
for mark in marks:
    if mark> average:
       count +=1

#Grades
grades= {
   "A":0,
   "B":0,
   "C":0,
   "D": 0}

for mark in marks:
   if mark>=90:
      grades["A"] +=1
   elif mark>=75:
    grades["B"] +=1
   elif mark>=60:
     grades["C"] +=1
   else:
      grades["D"]+=1

print("\n Student Peerformance Report")
print("Highest Marks:",highest)
print("Lowest marks:", lowest)
print("Average marks:",average)
print (" Above Average Students:",count)
print ("\nGrade distribution: ")
for grade, value in grades.items():
   print(f"{grade}: {value}")
      
````

````python
# Write your solution here
#print
face = [
    ["R", "R", "G"],
    ["W", "R", "G"],
    ["B", "B", "Y"]
]

#print
for row in face:

    for value in row:
       print(value, end=" ")
    print()
# counting the frequency
counts = {}
for row in face:
  for color in row:

    if color in counts:

        counts[color] += 1

    else:

        counts[color] = 1

print(counts)
# Replace 
for row in face:

    for index in range(len(row)):

        if row[index] == "R":

            row[index] = "X"

print(face)
#most common colour 
common= 1
count=0
for color in counts:
   if counts[color]>count:
      count= counts[color]
      common = color
print ("most common color:" , common)

#identical rows
found= False
for row in face:
    if row[0]== row[1] and row[1]==row[2]:
       found = True
       
if found:
     print ("\n Identical row")
else:
    print("\n No row is identical")
          


````
