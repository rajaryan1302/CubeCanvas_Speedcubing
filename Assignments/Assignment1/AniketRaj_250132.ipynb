## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
# Write your solution here

#input of student marks
student_marks = list(map(int, input().split()))

if len(student_marks) != 10:
    print("Please enter exactly 10 marks")

#Highest, Lowest and Average marks
Highest_marks = max(student_marks)
Lowest_marks = min(student_marks)
Average_marks = sum(student_marks) / len(student_marks)

count = 0
for num in student_marks:
    if num > Average_marks:
        count += 1

print("Highest Marks = ", Highest_marks)
print("Lowest Marks = ", Lowest_marks)
print("Average Marks = ", Average_marks)
print("Number of students scoring above average = ", count)

#Assigning Grades
for num in student_marks:
    if num >= 90:
        grade = "A"
    elif num >= 75:
       grade = "B"
    elif num >= 60:
       grade = "C"
    else:
       grade ="D"

#Grade Distribution
grade_distribution = {
    "A": 0,
    "B": 0,
    "C": 0,
    "D": 0
}

for mark in student_marks:
    if mark >= 90:
        grade_distribution["A"] += 1
    elif mark >= 75:
        grade_distribution["B"] += 1
    elif mark >= 60:
        grade_distribution["C"] += 1
    else:
        grade_distribution["D"] += 1

print("Grade Distribution =", grade_distribution)
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
# Write your solution here

face = [
    ["R", "R", "G"],
    ["W", "R", "G"],
    ["B", "B", "Y"]
]

#Printing Grid
for row in face:
    print(row)

#Counting frequency of every color
count_R, count_G, count_W, count_B, count_Y, count_O = 0, 0, 0, 0, 0, 0

for i in face:
    for j in i:
        if j == "R":
            count_R += 1
        elif j == "G":
            count_G += 1
        elif j == "W":
            count_W += 1
        elif j == "B":
            count_B += 1
        elif j == "Y":
            count_Y += 1
        elif j == "O":
            count_O += 1

print("Red Count = ", count_R)
print("Green Count = ", count_G)
print("White Count = ", count_W)
print("Blue Count = ", count_B)
print("Yellow Count = ", count_Y)
print("Orange Count = ", count_O)

#Determining Most Common Color
count = {}

for row in face:
    for color in row:
        count[color] = count.get(color, 0) + 1

most_common_color = max(count, key=count.get)

print("Most Common Color: ",most_common_color)

#Replacing "R" with "X"
for i in  range(len(face)):
    for j in range(len(face[i])):
        if face[i][j] == "R":
            face[i][j] = "X"


for row in face:
    print(row)

#Finding Rows with Identical Color
found = False

for i in range(len(face)):
    if all(x == face[i][0] for x in face[i]):
        print("All elements are identical in row", i)
        found = True

if not found:
    print("No row has identical colors")
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
# Write your solution here

sentence = input("Enter a sentence: ")

# 1. Total characters
total_characters = len(sentence)

# 2. Total words
total_words = len(sentence.split())

# 3. Count vowels
vowels = "aeiouAEIOU"
vowel_count = 0

for ch in sentence:
    if ch in vowels:
        vowel_count += 1

# 4. Frequency of every character
frequency = {}

for ch in sentence:
    frequency[ch] = frequency.get(ch, 0) + 1

# 5. Most frequently occurring character
most_frequent = max(frequency, key=frequency.get)

# 6. Palindrome check
processed = sentence.replace(" ", "").lower()
is_palindrome = processed == processed[::-1]

print("Total Characters =", total_characters)
print("Total Words =", total_words)
print("Total Vowels =", vowel_count)
print("Character Frequency =", frequency)
print("Most Frequent Character =", most_frequent)
print("Palindrome =", is_palindrome)
````

def modify_list(lst):

````python
# Write your solution here

#Immutable Example
def add_five(n):
    n += 5
    print("Inside function:", n)

x = 10

add_five(x)

print("Outside function:", x)

#Mutable Example
def add_item(lst):
    lst.append(4)

numbers = [1, 2, 3]

add_item(numbers)

print(numbers)

#Safe Function to modify a list
def safe_modify_list(lst):
    new_list = lst.copy()
    new_list.append(100)
    return new_list

values = [1, 2, 3]

new_values = safe_modify_list(values)

print("Original:", values)
print("Modified:", new_values)

````

# Mini Project (50 Marks): Speedcubing Club Management System

The Speedcubing Club wants a software system that can be reused by future coordinators to manage members, attendance and events.

## Class: Member

Each member should store:
- Name

Methods:

## Class: Event

## Class: Club

The Club class should:

## Bonus Features

Implement any two:

````python
# Mini Project Solution
# Mini Project Solution

class Member:
    def __init__(self, name, roll_no, branch, year, average):
        self.name = name
        self.roll_no = roll_no
        self.branch = branch
        self.year = year
        self.average = average

    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Roll No: {self.roll_no}")
        print(f"Branch: {self.branch}")
        print(f"Year: {self.year}")
        print(f"Average: {self.average}")

    def update_average(self, new_average):
        self.average = new_average


class Event:
    def __init__(self, event_name, date, venue):
        self.event_name = event_name
        self.date = date
        self.venue = venue
        self.registered_members = []

    def register_member(self, member):
        self.registered_members.append(member)

    def remove_member(self, roll_no):
        self.registered_members = [
            m for m in self.registered_members
            if m.roll_no != roll_no
        ]

    def display_participants(self):
        print(f"\nParticipants of {self.event_name}")
        for member in self.registered_members:
            print(member.name, member.roll_no)


class Club:
    def __init__(self):
        self.members = []
        self.events = []
        self.attendance = {}      # Bonus Feature 1

    def add_member(self, member):
        self.members.append(member)

    def remove_member(self, roll_no):
        self.members = [
            m for m in self.members
            if m.roll_no != roll_no
        ]

    def search_member(self, roll_no):
        for member in self.members:
            if member.roll_no == roll_no:
                return member
        return None

    def create_event(self, event):
        self.events.append(event)

    def display_all_events(self):
        print("\nAll Events")
        for event in self.events:
            print(event.event_name, event.date, event.venue)

    def display_all_members(self):
        print("\nAll Members")
        for member in self.members:
            member.display_details()
            print()

    def top_3_fastest(self):
        sorted_members = sorted(
            self.members,
            key=lambda x: x.average
        )

        print("\nTop 3 Fastest Cubers")

        for member in sorted_members[:3]:
            print(member.name, member.average)

    def average_cubing_time(self):
        if len(self.members) == 0:
            return 0

        total = sum(member.average for member in self.members)

        return total / len(self.members)

    def branch_wise_count(self):
        count = {}

        for member in self.members:
            count[member.branch] = (
                count.get(member.branch, 0) + 1
            )

        return count

    # Bonus Feature 1
    def mark_attendance(self, workshop_name, roll_no):
        if workshop_name not in self.attendance:
            self.attendance[workshop_name] = []

        self.attendance[workshop_name].append(roll_no)



    # Bonus Feature 4
    def search_by_branch(self, branch):
        result = []

        for member in self.members:
            if member.branch == branch:
                result.append(member)

        return result


````
