## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
marks = []
for i in range(10):
    mark = float(input(f"Enter marks of student {i + 1}: "))
    marks.append(mark)

highest = max(marks)
lowest = min(marks)
average = sum(marks) / len(marks)
above_average = 0
for mark in marks:
    if mark > average:
        above_average += 1

grade_distribution = {
    "A": 0,
    "B": 0,
    "C": 0,
    "D": 0
}
print(f"Highest marks: {highest}")
print(f"lowest marks: {lowest}")
print(f"average marks: {average}")
print(f" students above average: {above_average}")



for mark in marks:
    if mark >= 90:
        grade_distribution["A"] += 1
    elif mark >= 75:
        grade_distribution["B"] += 1
    elif mark >= 60:
        grade_distribution["C"] += 1
    else:
        grade_distribution["D"] += 1
for grade, count in grade_distribution.items():
    print(f"{grade}: {count}")
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
face = [
    ["R", "R", "G"],
    ["W", "R", "G"],
    ["B", "B", "Y"]
]
for row in face:
    print(row)

count = {}

for row in face:
    for color in row:
        if color not in count:
            count[color] = 1
        else:
            count[color] = count[color] + 1
print(count)

for row in face:
    for i in range(3):
        if row[i] == "R":
            row[i] = "X"
for row in face:
    print(row)

highest = 0
common = ""

for color in count:
    if count[color] > highest:
        highest = count[color]
        common = color
print(f"most common color is {common}")

same = False
for row in face:
    if row[0] == row[1] and row[1] == row[2]:
     same = True

if same==True:
    print("identical")
else:
    print("Not identical")
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
sentence = input("Enter a sentence: ")

characters = len(sentence)

words = sentence.split()
word_count = len(words)

vowel_count = 0

for ch in sentence.lower():
    if ch in "aeiou":
        vowel_count += 1

frequency = {}

for ch in sentence:
    if ch in frequency:
        frequency[ch] += 1
    else:
        frequency[ch] = 1

highest = 0
most_common = ""

for ch in frequency:
    if frequency[ch] > highest:
        highest = frequency[ch]
        most_common = ch

new_sentence = ""

for ch in sentence.lower():
    if ch != " ":
        new_sentence += ch

if new_sentence == new_sentence[::-1]:
    palindrome = True
else:
    palindrome = False

print("\nTotal Characters:", characters)
print("Total Words:", word_count)
print("Total Vowels:", vowel_count)

print("\nCharacter Frequency:")
print(frequency)

print("\nMost Frequent Character:", most_common)

if palindrome:
    print("The sentence is a palindrome.")
else:
    print("The sentence is not a palindrome.")
````

## Question 4: Functions, Scope and Mutability Challenge (Hard)

Consider the following code:

```python
def modify_number(x):
    x += 10

def modify_list(lst):
    lst.append(100)

num = 5
values = [1,2,3]

modify_number(num)
modify_list(values)

print(num)
print(values)
```

Answer the following:

1. Predict the output.
2. Explain why the outputs differ.
3. Explain mutable and immutable objects.
4. Write your own examples demonstrating:
   - One immutable type
   - One mutable type
5. Create a function that safely modifies a list without changing the original list.


````python
def safe_modify(lst):
    new_list = lst.copy()
    new_list.append(100)
    return new_list

values = [1, 2, 3]

new_values = safe_modify(values)

print("Original:", values)
print("Modified:", new_values)
````

The output would be
5
[1, 2, 3, 100]
because the num is immutable and list is mutable. Basically mutable objects are those whose value can be modified like the originality can be changed while immutable objects are those whose value cannot be chnaged rather a new variable is created within to accomodate the modification. examples include mutable- dictionary  immutable-float.
