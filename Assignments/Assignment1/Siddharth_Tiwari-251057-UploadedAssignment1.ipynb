## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
marks=[]
for i in range (10):
    num = int(input(f"Enter marks of student {i+1}: "))
    marks.append(num)
count = 0
for num in marks :
  if num > sum(marks)/10 :
    count= count+1
grade_dict = {"A": 0, "B": 0, "C": 0, "D": 0}
for num in marks :
  if num > 90:
    grade_dict["A"]+=1
  elif num > 75:
    grade_dict["B"]+=1
  elif num > 60:
    grade_dict["C"]+=1
  else :
    grade_dict["D"]+=1

print("Maximum marks obtained ",max(marks),"\nMinimum marks obtained ",min(marks),"\nAverage ",sum(marks)/10,"\nAbove Average ",count,"\nGrades",grade_dict)
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
face = [
    ["R", "R", "G"],
    ["W", "R", "G"],
    ["B", "B", "Y"]
]
for list in face :
  print(list)

colour_freq={"R":0,"G":0,"B":0,"Y":0,"W":0}

for list in face :
  for colour in list:
    if colour == "R":
      colour_freq["R"]+=1
    elif colour == "G":
      colour_freq["G"]+=1
    elif colour == "B":
      colour_freq["B"]+=1
    elif colour == "Y":
      colour_freq["Y"]+=1
    else :
      colour_freq["W"]+=1

print(colour_freq)
for i in range(len(face)):
    for j in range(len(face[i])):
        if face[i][j] == "R":
            face[i][j] = "X"
print()
for list in face :
  print(list)
most_common = max(colour_freq, key=colour_freq.get)
print("Most common colour:", most_common)
for row in face:
    if row[0] == row[1] == row[2]:
        found = True
        print("A row contains identical colors:", row)

if not found:
    print("No row contains all identical colors")
````

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
sentence = str(input("Enter sentence :"))
print(len(sentence))
print(len(sentence.split()))

vowels = "AEIOUaeiou"
count =0
for ch in sentence :
  if ch in vowels :
    count+=1
print("Number of vowels is :",count)
freq ={}
for ch in sentence :
  if ch in freq:
    freq[ch]+=1
  else :
    freq[ch]=1

most_common = ""

for ch in freq:
    if most_common == "" or freq[ch] > freq[most_common]:
        most_common = ch

print(most_common)
sentence= sentence.replace(" ","")
sentence = sentence.lower()
new_sentence= sentence[: :-1]
if new_sentence == sentence:
  print("Palindrome : True ")
else :
  print("Palindrome : False ")
````
