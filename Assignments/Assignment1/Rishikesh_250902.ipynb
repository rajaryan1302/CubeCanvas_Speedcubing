## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````
# Write your solution here
students = []
c = 0
for i in range(10):
    marks = int(input("Enter student's mark: "))
    students.append(marks)

print("Highest mark: ", max(students))
print("Lowest mark: ", min(students))
print("Average mark: ", sum(students)/len(students))
for i in range(10):
    if students[i] > sum(students)/len(students): c+=1
print("Number of students scoring above average mark: ", c)

dist = {'A': 0, 'B': 0, 'C': 0, 'D': 0}
for i in range(10):
    if students[i] >= 90:
        dist['A'] += 1
    elif students[i] >= 75:
        dist['B'] += 1
    elif students[i] >= 60:
        dist['C'] += 1
    else:
        dist['D'] += 1

print(dist)
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````
# Write your solution here
face = [
    ["R", "R", "G"],
    ["W", "R", "G"],
    ["B", "B", "Y"],
]

for row in face:
    for c in row:
        print(c, end = " ")
    print()

dist = {}
for row in face:
    for c in row:
        if c in dist:
            dist[c] += 1
        else:
            dist[c] = 1
print(dist)

for row in face:
    for i in range(len(row)):
        if row[i] == "R":
            row[i] = "X"

for row in face:
    for c in row:
        print(c, end = " ")
    print()

val = list(dist.values())
for k, v in dist.items():
    if v == max(val):
        print(k)

r1, r2, r3 = face
if r1 == r2 == r3:
    print("All rows are identical")
if r1 == r2:
    print("Row 1 and 2 are identical")
elif r2 == r3:
    print("Row 2 and 3 are identical")
elif r1 == r3:
    print("Row 1 and 3 are identical")
else:
    print("None of the rows are identical")
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````
# Write your solution here
sentence = input("Enter the sentence: ")

c_c = 0
c_w = 0
f_characters = {}
c_v = 0

for c in sentence:
    c_c += 1
print("Number of characters: ", c_c)
for c in sentence.split():
    c_w += 1
print("Number of words: ", c_w)
for c in sentence:
    if c.lower() in ('a', 'e', 'i', 'o', 'u'):
        c_v += 1
print("Number of vowels: ", c_v)

for c in sentence.lower():
    if c in f_characters:
        f_characters[c] += 1
    else:
        f_characters[c] = 1

print(f_characters)

val = list(f_characters.values())
for k, v in f_characters.items():
    if v == max(val):
        print("The most frequent character is ", k)

text = sentence.replace(" ", "").lower()

if text == text[::-1]:
    print("Palindrome: True")
else:
    print("Palindrome: False")
````

# Mini Project (50 Marks): Speedcubing Club Management System

The Speedcubing Club wants a software system that can be reused by future coordinators to manage members, attendance and events.

## Class: Member

Each member should store:
- Name
- Roll Number
- Branch
- Year
- Cubing Average

Methods:
- `display_details()`
- `update_average()`

## Class: Event

Each event should store:
- Event Name
- Date
- Venue
- List of Registered Members

Methods:
- `register_member()`
- `remove_member()`
- `display_participants()`

## Class: Club

The Club class should:

- Store all members.
- Store all events.
- Add a member.
- Remove a member.
- Search for a member using roll number.
- Create an event.
- Display all events.
- Display all members.
- Display the top 3 fastest cubers in the club.
- Find average cubing time of all members.
- Count members branch-wise.

## Bonus Features

Implement any two:

1. Attendance tracking for workshops.
2. Member leaderboard based on cubing averages.
3. Event-wise participant statistics.
4. Search members by branch.
5. Export member information in a readable text format.

## Reflection Questions

Answer the following:

1. Where did you use encapsulation?
2. Which classes interact with each other and why?
3. Which variables are instance variables?
4. Which variables could be class variables?
5. What modifications would be required to reuse your system for next year?


````
# Mini Project Solution

class Member:
    def __init__(self, name, r_no, branch, yr, avg):
        self.name = name
        self.r_no = r_no
        self.branch = branch
        self.yr = yr
        self.avg = avg

    def display_details(self):
        print("Name: ", self.name)
        print("Roll No: ", self.r_no)
        print("Branch: ", self.branch)
        print("Year: ", self.yr)
        print("Cubing average: ", self.avg)

    def update_average(self, n_avg):
        self.avg = n_avg

class Event:
    def __init__(self, name, date, venue):
        self.name = name
        self.date = date
        self.venue = venue
        self.reg_mem = []

    def register_member(self, member):
        self.reg_mem.append(member)

    def remove_member(self, member):
        self.reg_mem.remove(member)

    def display_participants(self):
        for member in self.reg_mem:
            print(member, end = " ")

class Club(Member, Event):
    def __init__(self):
        self.members = []
        self.events = []

    def add_member(self, member):
        self.members.append(member)

    def remove_member(self, member):
        self.members.remove(member)

    def search_member(self, r_no):
        for member in self.members:
            if member.r_no == r_no:
                print(member.name)

    def create_event(self, event):
        self.events.append(event)

    def display_all_members(self):
        for member in self.members:
            print(member.name, end = " ")
        print()

    def display_all_events(self):
        for event in self.events:
            print(event.name, end = " ")
        print()

    def top_3_avg(self):
        members_copy = self.members[:]

        for i in range(len(members_copy)):
            for j in range(i+1, len(members_copy)):
                if members_copy[i].avg > members_copy[j].avg:
                    members_copy[i], members_copy[j] = members_copy[j], members_copy[i]

        print("Top 3 Fastest Cubers:")

        for member in members_copy[:3]:
            print(member.name, "-", member.avg)

    def avg_cubing_time(self):
        total = 0
        for member in self.members:
            total += member.avg

        return round(total / len(self.members), 2)

    def branchwise_count(self):
        count = {}

        for member in self.members:
            if member.branch in count:
                count[member.branch] += 1
            else:
                count[member.branch] = 1

        for branch, count in count.items():
                print(branch, ":", count)

    def search_by_branch(self, branch):
        for member in self.members:
            if member.branch == branch:
                print(member.name)

    def leaderboard(self):
        members_copy = self.members[:]

        for i in range(len(members_copy)):
            for j in range(i + 1, len(members_copy)):
                if members_copy[i].avg > members_copy[j].avg:
                    members_copy[i], members_copy[j] = members_copy[j], members_copy[i]

        print("Leaderboard:")
        rank = 1
        for member in members_copy:
            print(rank, ".", member.name, "-", member.avg)
            rank +=1

club = Club()

#Example

m1 = Member("Rohan", "250001", "CSE", 2, 12.4)
m2 = Member("Aman", "250002", "EE", 2, 9.8)
m3 = Member("Priya", "250003", "PHY", 1, 11.2)
m4 = Member("Sneha", "250004", "CSE", 3, 8.5)

club.add_member(m1)
club.add_member(m2)
club.add_member(m3)
club.add_member(m4)

event1 = Event("CubeCanvas Workshop",
               "2-06-2026",
               "MPR")

event1.register_member(m1)
event1.register_member(m2)

club.create_event(event1)

club.display_all_members()

club.display_all_events()

club.top_3_avg()

print("Average Cubing Time:",
      club.avg_cubing_time())

club.branchwise_count()

club.search_by_branch("CSE")

club.leaderboard()
````
