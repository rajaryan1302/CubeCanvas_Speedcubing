```python
# Question 4: Functions, Scope and Mutability Challenge

# Given Code
def modify_number(x):
    x += 10

def modify_list(lst):
    lst.append(100)

num = 5
values = [1, 2, 3]

modify_number(num)
modify_list(values)

print(num)      # Output: 5
print(values)   # Output: [1, 2, 3, 100]

"""
1. Predicted Output:
5
[1, 2, 3, 100]

2. Explanation:
The outputs differ because integers are immutable while lists are mutable.
modify_number() changes only the local copy of x, so num remains unchanged.
modify_list() modifies the original list object using append(), so values changes.

3. Mutable and Immutable Objects:
Immutable objects cannot be changed after creation.
Examples: int, float, str, tuple

Mutable objects can be changed after creation.
Examples: list, dict, set

4. Examples:

Immutable Type Example:
"""
x = 10
x += 5
print("Immutable Example:", x)  # 15

"""
Mutable Type Example:
"""
numbers = [1, 2, 3]
numbers.append(4)
print("Mutable Example:", numbers)  # [1, 2, 3, 4]

"""
5. Function that safely modifies a list without changing the original list:
"""

def safe_modify_list(lst):
    new_list = lst.copy()
    new_list.append(100)
    return new_list

original = [1, 2, 3]
modified = safe_modify_list(original)

print("Original List:", original)
print("Modified List:", modified)
```
