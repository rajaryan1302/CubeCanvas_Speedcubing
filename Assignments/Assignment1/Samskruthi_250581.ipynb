## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
marks = [20, 23, 49, 56, 73, 98, 34, 57, 84]
print(marks)
# highest and lowest marks
highest = max(marks)
lowest = min(marks)
print(highest)
print(lowest)
# average marks
average = sum(marks)/len(marks)
print(average)
#number of students above average
above_average = 0
for mark in marks:
    if mark > average:
        above_average += 1
print(above_average)
#assign grades
grades = []
for m in marks:
  if m >= 90:
    grade = ("A")
  elif m >= 75:
    grade = ("B")
  elif m >= 60:
    grade = ("C")
  else:
    grade = ("D")
    print(above_average)
# grade distrubution using dictionary
grade_count = {"A": 0, "B": 0, "C": 0, "D": 0}
for g in grades:
  grade_count[g] += 1
print(grade_count)
````

````python
# Write your solution here
face = [
    ["R", "R", "G"],
    ["W", "R", "G"],
    ["B", "B", "Y"]
]
#print the grid neatly
for  row in face :
  for value in row :
    print (value , end =" ")
  print ()
#replace every occurance of R-->X
for row in face :
  for index in range (len(row)):
    if row [index] =='R':
       row [index]='X'
print (face )
freq = {}
for row in face :
  for color in row :
    if color in freq:
       freq [color]+=1
    else :
       freq[color] = 1
print (freq)
most_common = max( freq , key =freq.get)
print("Most common color;", most_common)
print("count:", freq[most_common])

found = False
# assume that no rows have common color at the start
for row in face:
   if row[0] == row [1] and row [1] == row[2]:
      found = True
      break
print (found)
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
# Write your solution here
sentence = "NO lemon, no melon."
# count total characters
char_count = len(sentence)
print(char_count)
# count total words
word_count = len(sentence.split())
print(word_count)
# count vowelss
vowels = "aeiouAEIOU"
vowel_count = 0
for char in sentence:
  if char in vowels:
    vowel_count += 1
    print(vowel_count)
# frequency of every charecter
freq = {}
for char in sentence:
  freq[char] = freq.get(char, 0) + 1
print(freq)
#most frequently occuring character
most_freq_char = max(freq, key  = freq.get)
print(most_freq_char)
# palindromecheck
cleaned = ""
for char in sentence.lower():
  if char.isalnum():
    cleaned += char
is_palindrome = cleaned == cleaned[::-1]
print(is_palindrome)

````
